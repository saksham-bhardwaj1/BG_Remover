```python
## Importing Libraries
import rembg
from PIL import Image
import matplotlib.pyplot as plt

## Define Function
def rmbg(x):
    image=Image.open(x)
    plt.imshow(image)
    plt.axis('off')
    plt.title("Image before removing Background")
    plt.show()
    output=rembg.remove(image)
    output.save("output.png")
    plt.imshow(output)
    plt.axis('off') 
    plt.title("Image after removing Background")
    plt.show()
    print("Your Image save successfully to your computer name output.png")

    
```
